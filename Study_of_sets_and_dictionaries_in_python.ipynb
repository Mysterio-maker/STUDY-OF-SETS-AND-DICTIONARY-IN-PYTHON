# Python frozenset

A frozenset is a built-in data type in Python that represents an immutable set.

 Immutable means: once a frozenset is created, its elements cannot be changed.

 Definition:-

A frozenset is an unchangeable (immutable) collection of unique elements in Python.

```python
3 # frozenset

x= frozenset({"Ferrari","Lamborghini","Bugatti"})
print(x)
print(type(x))
```

```python
# add and remove

x= frozenset({"Ferrari","Lamborghini","Bugatti"})
x.add("BMW")
```

```python
x= frozenset({"Ferrari","Lamborghini","Bugatti"})
x.remove("Ferrari")
```

```python
# set opeations using frozenset

A = frozenset({"Ferrari", "Lamborghini", "Bugatti"})
B = frozenset({"BMW", "Ferrari", "Lamborghini"})

print("Union:", A | B)
print("Intersection:", A & B)
print("Difference (A - B):", A - B)
print("Symmetric Difference:", A ^ B)
```

```python
# Length of dictionary

thisdict={
    "brand":"BMW",
    "model":"M series",
    "year":2000
}

print(len(thisdict))

```

```python
# Problem statement 4 = A dictionay used to store student marks and display maximum

student_marks = {
    "A": 70,
    "B": 65,
    "C": 75,
}

max_marks = max(student_marks.values())


topper = max(student_marks, key=student_marks.get)

print("Topper:", topper)
print("Maximum marks:", max_marks)
```
